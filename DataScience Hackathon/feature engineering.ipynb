```python
lags = [1, 2, 3]  # Define the lag values

for lag in lags:
    data_frame[f'total_related_ott_searches1_lag_{lag}'] = data_frame['total_related_ott_searches1'].shift(lag)
    data_frame[f'total_related_ott_searches2_lag_{lag}'] = data_frame['total_related_ott_searches2'].shift(lag)
    data_frame[f'total_related_ott_searches3_lag_{lag}'] = data_frame['total_related_ott_searches3'].shift(lag)
```

```python
data_frame.head(12)
```

```python
# Check for missing values in lag features and target variable
lag_columns = ['total_related_ott_searches1_lag_1', 'total_related_ott_searches2_lag_1', 'total_related_ott_searches3_lag_1',
               'total_related_ott_searches1_lag_2', 'total_related_ott_searches2_lag_2', 'total_related_ott_searches3_lag_2',
               'total_related_ott_searches1_lag_3', 'total_related_ott_searches2_lag_3', 'total_related_ott_searches3_lag_3',
               'Status_Subscribed']

missing_values = data_frame[lag_columns].isnull().sum()
print(missing_values)
```

```python
data_frame.dropna(subset=lag_columns, inplace=True)
```
